Task 1:


Problem Statement

Create a simple webpage with a header, a main content section, and a footer. Apply a media query that
changes the background color of the header to blue when the screen width is less than 600px.

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <link rel="stylesheet" href="./problem01.css">
</head>
<body>
    <header>
        <div>Responsive Header</div>
    </header>
    <main>
        Main
    </main>
    <footer>
        Footer
    </footer>
</body>
</html>

CSS File


* {
    margin: 0;
    padding: 0;
}
header {
    height: 100px;
    display: flex;
    justify-content: center;
    align-items: center;
    background-color: gray;
    font-size: 25px;
    font-weight: 900;
}
@media screen and (width <= 600px){
    header {
        background-color: blue;
    }
}
```

Task 3:


Problem Statement

Create a navigation bar with five links. Apply a media query that converts the navigation into a vertical list for
screens smaller than 768px and also changes the background color to gray.


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <link rel="stylesheet" href="./problem03.css">
</head>
<body>
    <div class="container">
        <a href="#">Home</a>
        <a href="#">About</a>
        <a href="#">Service</a>
        <a href="#">Portfolio</a>
        <a href="#">Contact Us</a>
    </div>
</body>
</html>

CSS File

* {
    margin: 0;
    padding: 0;
}
.container {
    height: 80px;
    display: flex;
    justify-content: space-around;
    background-color: black;
    align-items: center;
}
div a{
    text-decoration: none;
    color: white;
}
@media screen and (max-width: 768px){
    .container {
        background-color: grey;
        flex-direction: column;
        height: 120px;
    }
}
```

Task 4:


Problem Statement

Create a grid based layout containing four grid items which arranges them as the screen size changes.


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <link rel="stylesheet" href="./problem04.css">
</head>
<body>
    <div class="container">
        <div id="box1">Box 1</div>
        <div id="box2">Box 2</div>
        <div id="box3">Box 3</div>
        <div id="box4">Box 4</div>
    </div>
</body>
</html>

CSS File

* {
    margin: 0;
    padding: 0;
}
.container{
    border: 2px solid black;
    margin: 15px;
    padding: 15px;
    display: grid;
    grid-template-columns: 1fr 1fr 1fr;
    gap: 10px;

}
.container div {
    height: 200px;
    width: 100%;
}
.container div:first-of-type {
    background-color: red;
    color: white;
    text-align: center;
    font-size: 25px;
}
.container div:nth-child(2){
    background-color: green;
    color: white;
    text-align: center;
    font-size: 25px;
}
.container div:nth-child(3){
    background-color: blue;
    color: white;
    text-align: center;
    font-size: 25px;
}
.container div:nth-last-child(1) {
    background-color: orange;
    color: white;
    text-align: center;
    font-size: 25px;
}
@media screen and (max-width: 768px) {
    .container {
        grid-template-columns: 1fr;
    }
    .container div {
        width: 100%;
    }
}
```
