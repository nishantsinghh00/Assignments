Q2. Create a class-based component called Counter with an initial state value of count set to 0. Display the count value in the component. Implement two buttons, one to increment and one to decrement the count
value when clicked.



```python
// App.jsx

import Counter from "./Counter";

function App(){
   return(
    <Counter />
   );
}

export default App;


// class Counter.jsx

import { Component } from "react";

class Counter extends Component{
    state={
        count:0,
    }

    increment =()=>{
        this.setState({
            count:this.state.count+1,
        })
    }
    decrement = ()=>{
        this.setState({
            count:this.state.count-1,
        })
    }
    render(){
        return(
            <>

               <div className="bg-gray-500 flex flex-col mx-auto mt-[80px] h-[150px] w-[300px] gap-[40px] pl-[20px]">
                 <div className="text-xl text-white font-bold mt-[10px]">Counter: <span>{this.state.count}</span></div>
                 <div className="flex gap-[50px]">
                    <button className="w-[90px] h-[40px] bg-green-600 text-white" onClick={this.increment}>Increment</button>
                    {this.state.count>0 && <button className="w-[90px] h-[40px] bg-red-500 text-white" onClick={this.decrement}>Decrement</button>}
                 </div>
               </div>
            </>
        );
    }
}

export default Counter;
```

Q3. Build a class-based component called ToggleButton. Initially, the component should display a button with the text "OFF." When the button is clicked, it should toggle between "ON" and "OFF" each time it's clicked.

```python
// App.jsx

import ToggelButton from "./ToggelButton";

function App(){
    return(
        <ToggelButton />
    );

}

export default App;

// class Toggel.jsx


import { Component } from "react";

class ToggelButton extends Component{
    state={
        toggel : false,
    }
    btnHandler =()=>{
        if(this.state.toggel === false){
            this.setState({
                toggel: this.state.toggel=true,
            })
        }else{
            this.setState({
                toggel: this.state.toggel=false,
            })
        }
    }

    render(){
        return(
            <>
                <div className="h-[150px] w-[250px] bg-gray-500 mx-auto mt-[50px] flex flex-col text-3xl text-white font-bold justify-center items-center">
                    Toggel Button
                    <button className="h-[50px] w-[80px] bg-white border-red-600 rounded-md border-[2.5px] mt-5 text-black" onClick={this.btnHandler}>{this.state.toggel===false?"OFF":"ON"}</button>
                </div>

            </>
        );
    }
}

export default ToggelButton;
```

Q4. Create a class-based component Todolist that manages a list of tasks. Implement an input field for adding new tasks, a button to add a task to the list, and a list displaying all tasks. Use state to manage the list
of tasks and update it when a new task is added.


```python
// App.jsx

import ToDoList from "./ToDoList";

function App(){
   return(
    <ToDoList />
   );
}

export default App;


// class ToDoList.jsx


import { Component } from "react";

class ToDoList extends Component{
   state={
    taskArray:[],
    inputValue:"",
   }
   inpHandler=(e)=>{
    this.setState({
        inputValue : this.state.inputValue=e.target.value,
    })
   }
   btnHandler=()=>{
    this.setState({
        taskArray: this.state.taskArray= [...this.state.taskArray,this.state.inputValue],
        inputValue:this.state.inputValue="",
    })
   }


    render(){
        return(
            <div className="bg-gray-500 h-[600px] w-[500px] mx-auto mt-[80px] rounded-md flex flex-col px-5">
                <p className="text-3xl text-white font-bold mt-5 text-center">To-Do-List</p>
                <div className="flex p-3 gap-[40px]">
                    <div><input placeholder="Add Task" className="h-[40px] w-[300px] bg-white text-gray-400 p-1 rounded-md" value={this.state.inputValue} onChange={this.inpHandler} type="text"/></div>
                    <div><button className="bg-black text-white h-[40px] w-[80px] rounded-md cursor-pointer" onClick={()=>{
                        this.state.inputValue != "" && this.btnHandler();
                    }}>ADD</button></div>
                </div>
                <div className="border-2 bg-black w-full rounded-md mt-4"></div>
                <div className="flex flex-col mt-5 gap-2">
                    {this.state.taskArray?.map((task,index)=>{
                        return(
                           <p key={index} className="text-xl text-white font-bold"><span className="text-white">{index+1}. </span>{task}</p>
                        );
                    })}
                </div>
            </div>
        );
    }
}

export default ToDoList;
```
