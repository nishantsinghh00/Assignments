Task 3:


Problem Statement

Explain the use of grid-auto-row and grid-auto-column using code examples.
 -  grid-auto-row:
    -  It defines the height of row which row is not defined in grid-template-row.
    -  If we give only two row height in grid-template-row(100px 100px) and grid-auto-row(50px) then after two row all new  row height is 50px.
 -  grid-auto-column:
    -  It works same as grid-auto-row. It defines the width of column which column is not defined in grid-template-row.
    -  If we give only two column width in grid-template-column(100px 100px) and grid-auto-column(50px) then after two column all new column width is 50px.

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <style>
        #container {
            width: 300px;
            height: 500px;
            display: grid;
            grid-template-rows:100px 100px ;
            grid-auto-rows: 50px;
            grid-template-columns: 1fr 1fr;
            /* grid-auto-flow: column; */
            grid-auto-columns: 50px;
            border: 2px solid black;
            padding: 25px;
            gap: 5px;
        }
        #container div {
            height: 100%;
            width: 100%;
            background-color: aqua;
        }
    </style>
</head>
<body>
    <div id="container">
        <div>1</div>
        <div>2</div>
        <div>3</div>
        <div>4</div>
        <div>5</div>

    </div>
</body>
</html>
```

Problem Statement

Explain the difference between justify-items and justify-self using code examples.
 -  justify-items:
    - It written in grid continer.
    - It applies on all grid iteams inside thier cell.
 -  justify-self:
    - It written in individual grid iteam.
    - It applies on specific grid iteam inside thier cell.

 -  Both do same work Align All iteam Horizontaly inside grid iteam cell.    

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <style>
        #container {
            width: 600px;
            height: 900px;
            display: grid;
            grid-template-rows:100px 100px 100px ;
            /* grid-auto-rows: 50px; */
            grid-template-columns: 100px 100px;
            /* grid-auto-flow: column; */
            /* grid-auto-columns: 50px; */
            border: 2px solid black;
            padding: 25px;
            gap: 5px;
            justify-content: end;
            align-content: center;
            justify-items: end;
            align-items: end;
        }
        #container div {
            height: 50px;
            width: 50px;
            background-color: aqua;
        }
        #container div:first-of-type {
            justify-self: center;
        }
    </style>
</head>
<body>
    <div id="container">
        <div>1</div>
        <div>2</div>
        <div>3</div>
        <div>4</div>
        <div>5</div>

    </div>
</body>
</html>
```
