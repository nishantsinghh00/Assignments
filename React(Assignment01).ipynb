Problem Statement

Q1. Explain the potential SEO challenges and solutions associated with Single Page Applications (SPAs).

 - SEO Challenges:
    - SPA use javascript to load full content after page starts so the Serach Engines can't see content.
    - SPA loads big javascript bundle before showing anything that makes website slow, bad user exprince.
    - Meta tag does not update so all pages look indentical to search engine.
    - No unique pages(All same url) in SPA sometimes URL does not change so search engine can't find our page.
  - SEO Solutions:
    - Dynamic meta tags use libraries to update  meta tags when route changes.
    - Speed optimization Split big javascript into small parts(code splitting).
    - Make sure each page has its own URL.

Problem Statement

Explain the key differences and use cases between React's Client-Side Rendering (CSR) and Server-Side
Rendering (SSR). Provide examples of scenarios where each approach is advantageous, and discuss the
challenges associated with using React in both contexts.

 - CSR(Client side rendering):
   - In CSR the srever sends bare minimum html with root div and sends bundel of javascript.
   - React takes over the browser and bulid UI dynamicaly, attach event listners.
   - the page is almost empty first then browser download javascript bundel then react renders UI to client.
 - Advantage of CSR:
   - Perfect for SPAs(single page application).
   - Once the loaded all things switching page feels instantly.
 - Challenges of CSR:
   - Search engine might struggel to find index page.
   - Users wait until all javascript bundel downloaded.
   - More CPU usage in browser.
 - SSR(Server-Side rendering):
   - In SSR the react app is pre-rendered into HTML on the server and sent to the browser.
   - The client still need to download React js.
   - In SSR user sees fully rendered HTML page instantly,then React js takes over for interactivity.
 - Advantage of SSR:
   - Search engine sees real content immediately.
   - User sees page content quickly improves UX.
 - Challenges of SSR:
   - Browser still needs to download React js to attach event listner after html is served.
   - Each request needs rendering work.
   - Requires complex framework like next.js or custom node.js server.
