Task 2:


Problem Statement

Explain the difference between Absolute and Relative positioning.

  -  Position Relative:
    -  When we use Position Relative the element to its normal position in document flow.
    -  The element still takes space in layout.
    -  We can use this for moving element left,right,top,buttom.
  -  Position Absolute:
    -  When we use Position Absolute the element removed from to its normal position in document flow and positioned relative to the nearest positioned ancestor.If ancestor is not with positioned then its relative to HTML.
    -  The element does not takes space in layout.
    - We can use this when we want to remove element from normal document flow and moving element left,right,top,buttom.
