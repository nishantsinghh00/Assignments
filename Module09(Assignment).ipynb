Q1. Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When would you choose to use one over the other?
  - Flexbox and Grid both are very powerfull tool in CSS for creating layouts but they are designed for different use cases.
  - Flexbox:
    - Flexbox is use for one dimensional(1D) layout.Works either row or coloumn but not both at same time.
    - Best for layout within component or for smaller UI elemnts.
    - Use cases:
      - Navigation bar.
      - Aligning buttom.
      - Simple list or iteam in single direction.
    - Grid:
      - Grid is use for two dimensional(2D) layout.Works on both row and coloumn simultaneously.
      - Most powerfull for creating full page or complex layout.
      - Use cases:
        - Page layout with header,footer,aside and content.
        - Dashboards.
        - Rseponsive design with clearly defined layouts areas.

Q2. Explain the role of the following key properties in the Flexbox layout model
  - justify-content
  - align-items
  - gap
  - flex-direction
  - flex-wrap
    - 1. justify-content: Controls the element along with main axis(which row by default).
    - 2. align-items: Controls the element along with cross axis(which coloumn by default).
    - 3. gap: Controls the spacing between flex items.
    - 4. flex-direction: Define the direction of main aixs and another axis is cross axis.
    - 5.flex-wrap: Controls the element whether they wrap if they don't fit in one line.
